Task - 3 - Outlier Detection:

```python
# 1.Identify outlier countries in South Asia based on their Score and GDP per Capita
from google.colab import drive
drive.mount('/content/drive/')
%cd /content/drive/MyDrive/Files

import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('south_asian_countries.csv')
q1_score=df['score'].quantile(0.25)
q3_score=df['score'].quantile(0.75)
iqr_score=q3_score-q1_score
lower_score=q1_score-1.5*iqr_score
upper_score=q3_score-1.5*iqr_score

q1_gdp=df['Log GDP per capita'].quantile(0.25)
q3_gdp=df['Log GDP per capita'].quantile(0.75)
iqr_gdp=q3_gdp-q1_gdp
lower_gdp=q1_gdp-1.5*iqr_gdp
upper_gdp=q3_gdp-1.5*iqr_gdp

# Define outliers using the 1.5 × IQR rule.
outliers=df[(df['score']<lower_score) | (df['score']>upper_score) | (df['Log GDP per capita']<lower_gdp) | (df['Log GDP per capita']>lower_gdp)]

# 3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.
plt.scatter(df['Log GDP per capita'], df['score'], color='purple', label='Non-Outliers')
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='brown', label='Outliers')

# 4.Discussion and Ploting
plt.title("Outlier Detection: GDP vs Score")
plt.xlabel("GDP per Capita")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(outliers[['Country name', 'score', 'Log GDP per capita']])
```

```python
# 6. Outlier Detection:
# Identify outlier countries in both regions based on Score and GDP per Capita.
# Plot these outliers and discuss their implications.

# Define a function to calculate outliers using the 1.5 × IQR rule
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)  # 25th percentile
    Q3 = df[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Detect outliers for Middle East
me_outliers_score, me_lower_score, me_upper_score = detect_outliers(middle_east_data, "score")
me_outliers_gdp, me_lower_gdp, me_upper_gdp = detect_outliers(middle_east_data, "Log GDP per capita")

# Detect outliers for South Asia
sa_outliers_score, sa_lower_score, sa_upper_score = detect_outliers(south_asia_data, "score")
sa_outliers_gdp, sa_lower_gdp, sa_upper_gdp = detect_outliers(south_asia_data, "Log GDP per capita")

# Scatter plots for visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)
fig.suptitle("Outliers in Score and GDP per Capita", fontsize=16)

# Middle East scatter plot
sns.scatterplot(ax=axes[0], x="Log GDP per capita", y="score", data=middle_east_data, label="Middle East", color="blue")
sns.scatterplot(ax=axes[0], x="Log GDP per capita", y="score", data=me_outliers_score, label="Outliers (Score)", color="red")
sns.scatterplot(ax=axes[0], x="Log GDP per capita", y="score", data=me_outliers_gdp, label="Outliers (GDP)", color="orange")
axes[0].set_title("Middle East")
axes[0].set_xlabel("Log GDP per capita")
axes[0].set_ylabel("Score")
axes[0].legend()

# South Asia scatter plot
sns.scatterplot(ax=axes[1], x="Log GDP per capita", y="score", data=south_asia_data, label="South Asia", color="green")
sns.scatterplot(ax=axes[1], x="Log GDP per capita", y="score", data=sa_outliers_score, label="Outliers (Score)", color="red")
sns.scatterplot(ax=axes[1], x="Log GDP per capita", y="score", data=sa_outliers_gdp, label="Outliers (GDP)", color="orange")
axes[1].set_title("South Asia")
axes[1].set_xlabel("Log GDP per capita")
axes[1].legend()

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
